# Dataset Description

==> Province_State - The name of the State within the USA.

==> Country_Region - Country, region or sovereignty name..

==> Last_Update - The most recent date the file was pushed.(MM/DD/YYYY HH:mm:ss (24 hour format, in UTC)).

==> Lat - Latitude.

==> Long_ - Longitude.

Dot locations on the dashboard. All points (except for Australia) shown on the map are based on geographic centroids, and are 

not representative of a specific address, building or any location at a spatial scale finer than a province/state. Australian 

dots are located at the centroid of the largest city in each state.

==> Confirmed - Aggregated confirmed case count for the state.Confirmed cases include presumptive positive cases and probable 

cases, in accordance with CDC guidelines as of April 14.

==> Deaths - Aggregated Death case count for the state.Death totals in the US include confirmed and probable, in accordance with CDC guidelines as of April 14.

==> Recovered - Aggregated Recovered case count for the state.Recovered cases outside China are estimates based on local media 

reports, and state and local reporting when available, and therefore may be substantially lower than the true number. US state-

level recovered cases are from COVID Tracking Project.

==> Active - Aggregated confirmed cases that have not been resolved (Active = Confirmed - Recovered - Deaths).

==> FIPS - Federal Information Processing Standards code that uniquely identifies counties within the USA.

==> Incident_Rate - confirmed cases per 100,000 persons.

==> People_Tested - Total number of people who have been tested.

==> People_Hospitalized - Total number of people hospitalized.

==> Mortality_Rate - Number recorded deaths * 100/ Number confirmed cases.

==> UID - Unique Identifier for each row entry.

==> ISO3 - Officialy assigned country code identifiers.

==> Testing_Rate - Total number of people tested per 100,000 persons.The "total test results" is equal to "Total test results 

(Positive + Negative)" from COVID Tracking Project.

==> Hospitalization_Rate - Total number of people hospitalized * 100/ Number of confirmed cases.

==> Admin2: County name. US only.

# Removing Outliers

Removing outliers using Interquartile Range

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
(df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))
df1 = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
df1
```

```python
print(df.shape,'\t',df1.shape)
```

58 rows has reduced to 30.So,outliers are removed.

```python
df = df1
```
